### [Problem 4] Examining the iris dataset itself

The Fisher's Iris dataset is a widely used dataset in the field of machine learning and statistics. It was introduced by the British biologist and statistician Ronald A. Fisher in his 1936 paper "The use of multiple measurements in taxonomic problems" as an example of discriminant analysis. The dataset consists of 150 observations of iris flowers, with 50 observations for each of the three species: setosa, versicolor, and virginica.

Here are the key attributes of the dataset:

1. Features (Measurements):
   - Sepal Length (in centimeters)
   - Sepal Width (in centimeters)
   - Petal Length (in centimeters)
   - Petal Width (in centimeters)

2. Target Variable (Response):
   - Species:
     - Setosa
     - Versicolor
     - Virginica

3. Number of Observations (Samples):
   - 50 samples of each species, totaling 150 observations.

4. Purpose:
   - The dataset is commonly used for statistical analysis, machine learning, and data visualization tasks. It is a classic dataset for practicing classification algorithms and exploring techniques in data analysis.

5. Distribution:
   - The measurements in each class (species) are normally distributed, making it a useful dataset for statistical analysis.

### [Problem 6] Creating a diagram

```python
%matplotlib inline
import matplotlib.pyplot as plt

# Make a pie chart of the number of samples for each label (display up to a percentage)
labels = data.target_names
sizes = [len(df.loc [df['Species'] == percentage]) for percentage in range( len(data.target_names) )]
fig1, ax1 = plt.subplots()
ax1.pie(sizes, labels = labels, autopct= '%1.1f%%', shadow = True, startangle = 90)
ax1.axis('equal')
plt.show()
```

```python
# Select features one by one and visualize the distribution of data for each label using a box plot
features = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']

species = data.target_names

for feature in features:

  a_list=[]

  for a in range(len(species)):
        species_found = df.loc[df['Species'] == a]
        a_list.append(species_found[feature ])
  fig1, ax1 = plt.subplots()
  ax1.set_title(feature)
  ax1.boxplot(a_list)
```

```python
# Select features one by one and visualize the distribution of data for each label using a violin plot
features = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']

species = data.target_names

for feature in features:

  a_list=[]

  for a in range(len(species)):
        species_found = df.loc[df['Species'] == a]
        a_list.append(species_found[feature ])
  fig1, ax1 = plt.subplots()
  ax1.set_title(feature)
  ax1.violinplot(a_list)
```

### Box plots and violin plots display similar plots. Explain how they differ and what advantages each has.

**Box Plot:**
- Shows minimum, first quartile, median, third quartile, and maximum.
- Compact representation of central tendency and spread.
- Highlights outliers effectively.
- Suitable for comparing multiple groups.

**Violin Plot:**
- Combines box plot with kernel density plot.
- Represents summary statistics and data distribution.
- Offers detailed view of distribution shape.
- Effective for comparing distributions with different shapes.

**Differences**:
- **Shape:** Box plots are simpler; violin plots show distribution shape.
- **Complementarity:** Box plots for central tendency and outliers; violin plots for overall distribution.
- **Interpretability:** Box plots are straightforward; violin plots offer more detail.



### [Problem 7] Confirming the relationship between features

```python
# Make scatter plots of all combinations (six), such as choosing two from four features, color-coded by type.
figure = plt.figure(figsize = (15, 5))
species = data.target_names
x_features = ['sepal_length', 'sepal_width']
y_features = ['petal_length', 'petal_width']
colors = ['red','green','orange']

for x_feature in range(len(x_features)):
    for a in range(len(species)):
        species_found = df.loc[df['Species'] == a]
        ax1= figure.add_subplot(1,2,x_feature+1)
        ax1.scatter(species_found [x_features[x_feature]], species_found[y_features[x_feature]], color=colors[a], label = species)
    ax1.legend()
    ax1.set_title(x_features[x_feature])
    ax1.set_xlabel(x_features)
    ax1.set_ylabel(y_features)
```
